```python
from bs4 import BeautifulSoup
import re

# Function to clean and extract text from HTML content
def clean_html_to_text(html_content):
    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Extract text from the HTML soup object, filtering out script and style elements
    for script_or_style in soup(["script", "style"]):
        script_or_style.decompose()  # Remove these tags from the soup
    
    # Get text
    text = soup.get_text()
    
    # Break into lines and remove leading and trailing space on each
    lines = (line.strip() for line in text.splitlines())
    # Break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    # Drop blank lines
    text = '\n'.join(chunk for chunk in chunks if chunk)
    
    return text

# Function to write text to a file
def write_text_to_file(text, filename="output.txt"):
    with open(filename, "w", encoding="utf-8") as file:
        file.write(text)

# Example HTML content (you can replace this with html_content variable to parse different HTML)
html_file = 'Graham.txt'
file = open(html_file, "r")
html_content = file.read()
file.close()
# Cleaning and extracting text
cleaned_text = clean_html_to_text(html_content)

# Writing cleaned text to a file
write_text_to_file(cleaned_text, "professor_webpage_text.txt")

print("Text extracted and saved to professor_webpage_text.txt")
```

```python
import requests
from bs4 import BeautifulSoup

# Function to fetch HTML content from a URL
def fetch_html_content(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raises an HTTPError if the response status code is 4XX/5XX
        return response.text
    except requests.RequestException as e:
        print(f"Error fetching HTML content from {url}: {e}")
        return None

# Function to clean HTML and extract text with hyperlinks
def clean_html_to_text_with_links(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    
    def process_element(element):
        text_parts = []
        for part in element.descendants:
            if part.name == "a":
                href = part.get('href', '')
                text = part.get_text()
                text_with_link = f"{text} ({href})" if href else text
                text_parts.append(text_with_link)
            elif part.string and not part.name:
                text_parts.append(part.string)
        return "".join(text_parts)
    
    text = process_element(soup)
    lines = (line.strip() for line in text.splitlines())
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    cleaned_text = '\n'.join(chunk for chunk in chunks if chunk)
    
    return cleaned_text

def main():
    url = "https://lti.cs.cmu.edu/people/222228444/fernando-diaz"  # Replace with the actual URL you want to fetch
    html_content = fetch_html_content(url)
    
    if html_content:
        cleaned_text_with_links = clean_html_to_text_with_links(html_content)
        print(cleaned_text_with_links)

if __name__ == "__main__":
    main()
```

```python
from unstructured.partition.html import partition_html
from unstructured.partition.pdf import partition_pdf

import os

def process_pdf(self, elements: list = None):
    texts = []
    page_num = elements[0].metadata.page_number

    for element in elements:
        if element.metadata.page_number != page_num:
            texts.append("\n")
            page_num = element.metadata.page_number
        texts.append(element.text)

    return "\n".join(texts)

def parse_pdf(self, file_path: str = None, include_page_breaks: bool = False):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")

    elements = partition_pdf(
        filename=file_path,
        languages=["eng"],
        include_page_breaks=include_page_breaks,
    )

    self.pdf_elements = elements
    return elements

path = 'test_zzh/2207.10397.pdf'
```

```python
import PyPDF2

def pdf_to_txt(pdf_path, txt_path):
    with open(pdf_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        num_pages = len(pdf_reader.pages)

        with open(txt_path, 'w') as txt_file:
            for page_num in range(num_pages):
                page = pdf_reader.pages[page_num]
                text = page.extract_text()
                txt_file.write(text)

if __name__ == "__main__":
    pdf_path = "2207.10397.pdf"  # Replace with the path to your PDF file
    txt_path = "output.txt"   # Replace with the path where you want to save the text file
    pdf_to_txt(pdf_path, txt_path)
```

```python
import os
import shutil

def move_txt_files(source_dir, target_dir):
    """
    Move all .txt files from source_dir to target_dir.

    Args:
    - source_dir (str): The path to the directory from which to move the files.
    - target_dir (str): The path to the directory where the files will be moved.
    """
    for filename in os.listdir(source_dir):
        if filename.endswith('.txt'):
            source_path = os.path.join(source_dir, filename)
            target_path = os.path.join(target_dir, filename)
            shutil.move(source_path, target_path)
            print(f"Moved: {source_path} to {target_path}")

# Example usage
source_directory = '/home/stevenzhang/11711Homeworks/LTI_Neural_Navigator/data/Prof_papers'
target_directory = '/home/stevenzhang/11711Homeworks/LTI_Neural_Navigator/data/Prof_papers/2024-02-27/pdf'
move_txt_files(source_directory, target_directory)
```
